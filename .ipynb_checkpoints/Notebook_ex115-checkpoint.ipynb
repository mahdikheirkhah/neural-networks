The goal of this exercise is to learn to adapt the output layer to regression.
As a reminder, one of the reasons why the sigmoid is used in classification is because it contracts the output between 0 and 1 which is the expected output range for a probability (W2D2: Logistic regression). However, the output of the regression is not a probability.

In order to perform a regression using a neural network, the activation function of the neuron in the output layer has to be modified to the identity function. In mathematics, the identity function is: **f(x) = x**. In other words it means that it returns the input as is. The three steps become:

1. Each input is multiplied by a weight
   - x1 -> x1 \* w1
   - x2 -> x2 \* w2
2. The weighted inputs are added together with a bias b
   - (x1 _ w1) + (x2 _ w2) + b
3. The sum is passed through an activation function
   - y = f((x1 _ w1) + (x2 _ w2) + b)
   - The activation function is **the identity**
   - y = (x1 _ w1) + (x2 _ w2) + b

All other neurons' activation function **do not change**.

````python
class Neuron:
    def __init__(self, weight1, weight2, bias, regression):
        self.weights = np.array([weight1, weight2])
        self.bias = bias
        self.regression = regression

    def feedforward(self, x1, x2):
        inputs = np.array([x1, x2])
        z = np.dot(self.weights, inputs) + self.bias
        if(self.regression):
            y = z
        else:
            y = 1 / (1 + np.exp(-z))
        return y


class OurNeuralNetwork:

    def __init__(self, neuron_h1, neuron_h2, neuron_o1):
        self.h1 = neuron_h1
        self.h2 = neuron_h2
        self.o1 = neuron_o1

    def feedforward(self, x1, x2):
        h1_feedforward = self.h1.feedforward(x1, x2)
        h2_feedforward = self.h2.feedforward(x1, x2)
        y = self.o1.feedforward(h1_feedforward, h2_feedforward)
        return y
````

````python
neuron = Neuron(0,1,4, True)
print(neuron.feedforward(2,3))
````

Compute and return the output of the neural network for each of the students. Here are the weights and biases of the neural network:

```
    #replace regression by the right value
    neuron_h1 = Neuron(0.05, 0.001, 0, regression)
```

````python
neuron_h1 = Neuron(0.05, 0.001, 0, False)
neuron_h2 = Neuron(0.002, 0.003, 0, False)
neuron_o1 = Neuron(2,7,10, True)

NN = OurNeuralNetwork(neuron_h1, neuron_h2, neuron_o1)
results = np.array([NN.feedforward(12 , 15), NN.feedforward(10 , 9), NN.feedforward(18 , 18), NN.feedforward(13 , 14)])
print(results)
````

3. Compute the MSE for the 4 students.

````python
y_true = np.array([16, 10, 19, 16])
print(np.mean(np.square(results - y_true)))
````
