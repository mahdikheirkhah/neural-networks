````python
import numpy as np
from sklearn.metrics import log_loss
````


The goal of this exercise is to compute the log loss on the output of the forward propagation. The data used is the tiny dataset below.

| name | math | chemistry | exam_success |
| :--- | ---: | --------: | -----------: |
| Bob  |   12 |        15 |            1 |
| Eli  |   10 |         9 |            0 |
| Tom  |   18 |        18 |            1 |
| Ryan |   13 |        14 |            1 |

The goal of the network is to predict success at the exam given math and chemistry grades. The inputs are `math` and `chemistry` and the target is `exam_success`.

1. Compute and return the output of the neural network for each of the students. Here are the weights and biases of the neural network:

   ```
   neuron_h1 = Neuron(0.05, 0.001, 0)
   neuron_h2 = Neuron(0.02, 0.003, 0)
   neuron_o1 = Neuron(2,0,0)
   ```



````python
class Neuron:
    def __init__(self, weight1, weight2, bias):
        self.weights = np.array([weight1, weight2])
        self.bias = bias

    def feedforward(self, x1, x2):
        inputs = np.array([x1, x2])
        z = np.dot(self.weights, inputs) + self.bias
        y = 1 / (1 + np.exp(-z))
        return y


class OurNeuralNetwork:

    def __init__(self, neuron_h1, neuron_h2, neuron_o1):
        self.h1 = neuron_h1
        self.h2 = neuron_h2
        self.o1 = neuron_o1

    def feedforward(self, x1, x2):
        h1_feedforward = self.h1.feedforward(x1, x2)
        h2_feedforward = self.h1.feedforward(x1, x2)
        y = self.o1.feedforward(h1_feedforward, h2_feedforward)
        return y
````

````python
neuron_h1 = Neuron(0.05, 0.001, 0)
neuron_h2 = Neuron(0.02, 0.003, 0)   
neuron_o1 = Neuron(2,0,0)

NN = OurNeuralNetwork(neuron_h1, neuron_h2, neuron_o1)
bob = NN.feedforward(12, 15)
Eli = NN.feedforward(10, 9)
Tom = NN.feedforward(18, 18)
Ryan = NN.feedforward(13, 14)
print("bob : ", bob)
print("Eli : ", Eli)
print("Tom : ", Tom)
print("Ryan : ", Ryan)
````

2. Compute the log loss for the data given the output of the neural network with the 4 students.

````python
y_pred = np.array([bob, Eli, Tom, Ryan])
y_true = np.ones(4)
y_true[1] = 0
print(y_pred)
print(y_true)
print(log_loss(y_true, y_pred))
````
